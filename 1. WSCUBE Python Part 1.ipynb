#### implement a stack and Queue using a List Data type 
#### Stack in Python 
 - The stack is a linear data structure, where the data gets stored in sequential order (one by one)
 - Stores items in a Last-in-first-out(LIFO) or First-in-last-out(FILO) manner
 

```python
l=[]
while True:
    choice=eval(input('''
    1 Push Elements
    2 Pop Elements
    3 Peek Elements
    4 Display Stack
    5 Exit
    ''' ))
    if choice==1:
        cnt=1
        for i in range(int(input("How many elements you want to add: "))):
            l.append(input(f"enter the value{cnt}: "));
            cnt+=1
        print(l)
    elif choice==2:
        if len(l)==0:
            print("Empty Stack")
        else:
            print("Deleting the Last Element: ",l.pop())
            print(l)
    elif choice==3:
        print(f"Displaying the Last Element of the {l} is {l[-1]}")
    elif choice==4:
        print("Displaying the Stack: ",l)
    elif choice==5:
        print("Program Terminated")
        break
    else:
        print("Invalid Input")
```

# Pickle In Python
 - what is Pickle ?

The data or the  datatypes in python , if we want to store iit in the file, we use Pickle Library.
 - Serialize means storing and DeSerialize means reading.

## Storing data with Pickle


## pickling files

 - dump(): we store the data in the file in a serialize way.
 - load(): it reads the data which is in serialize form.

```python
import pickle
a=[10,20,30,40]
file=open("writedata.txt","wb")
pickle.dump(a,file)
file.close()
```

# Unpickling with Load()

With the help of load(), we can de-serialize the data and read it.

```python
import pickle 
file=open("writedata.txt","rb")
l=pickle.load(file)
print(l)
```
